# Encoding Data Very Important

## OrdinalEncoder

```python
# We use an Encoder to Convert Data from Text to numbers bcz Algorithms work on Numbers not on characters

from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car',as_frame=True).frame
```

```python
data
```

- here we can see that safety amd lug_boot columns are not in numerical form to perform ML
- so we will convert this to numeric

## *How do we know on which columns we have to apply encoder or not?*

-  your data should we be in Order for example there is an order in education that fist there is primary educatioin then scendoray and so on same in these columns there is an order small med high low mid high so we use *OrdinalEncoder* here

- for other columns like countries where there is no oder u cannot classify those columns on the bases of order or rank u use another encoder which is called *OneHotEncoder*


```python
columns_to_encode = ['lug_boot','safety']

encoder = OrdinalEncoder(
    categories = [
        ['small','med','big'],
        ['low', 'med','high']
    ]
)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

```python
data
```

- now u can see numbers in lug_boot and in safety columns

### We can also inverse the transform

```python
encoder.inverse_transform(data[columns_to_encode])
```

# OneHotEncoder

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data = fetch_openml('adult',as_frame=True).frame
```

```python
data
```

```python
data.occupation.value_counts()
```

```python
encoder = OneHotEncoder(handle_unknown = 'ignore', sparse_output = False)

encoded_values = encoder.fit_transform(data[['occupation','race']])
new_cols = encoder.get_feature_names_out(['occupation','race'])
```

```python
new_cols
```

```python
df_encoded = pd.DataFrame(encoded_values , columns = new_cols, index =data.index)
data_final = pd.concat(
    [data.drop(columns=['occupation','race']),df_encoded],
    axis=1
)
```
