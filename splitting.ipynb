```python
from sklearn.datasets import load_iris
from sklearn.model_selection import  train_test_split
```

```python
data=load_iris()
X,y = data.data,data.target
```

```python
#Stratify = “keep class balance same in every split”

X_train , X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify= y)
len(X_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# np.bincounts means It counts how many times each non-negative integer appears in an array.
# we use it in ml to check the imbalance 

counts = np.bincount(y_train)
positions = np.arange(3)

plt.bar(positions,counts)
plt.xticks(positions , data.target_names)
```

```python
counts = np.bincount(y_test)
positions = np.arange(3)

plt.bar(positions,counts)
plt.xticks(positions , data.target_names)
```

```python
# we use this when :- 
# Dataset is small
# You want more reliable evaluation
# You care about variance in results
# Generally we use train_test_split

from sklearn.model_selection import StratifiedShuffleSplit  # it gurantees to have a similar ratio of distribution
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)

for train_idx , test_idx in split.split(X,y):
    X_train ,X_test = X[train_idx], X[test_idx]
    y_train ,y_test = y[train_idx], y[test_idx]
```
